```python
#adding your Tile Layers
m = folium.Map(location=[47.809490, 13.055010])
folium.TileLayer('Stamen Terrain').add_to(m)
folium.TileLayer('Stamen Toner').add_to(m)
folium.TileLayer('Stamen Watercolor').add_to(m)
folium.TileLayer('cartodb positron').add_to(m)  # Corrected from 'cartodbpositron'
folium.TileLayer('cartodb dark_matter', attr='©OpenStreetMap, ©CartoDB').add_to(m)  # Added attribution
folium.LayerControl().add_to(m)
m
```

```python
# Load the data
url = 'https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv'
earthquake_data = pd.read_csv(url)

# Create a map
m = folium.Map(location=[0, 0], zoom_start=2)

# Add the heatmap layer
plugins.HeatMap(
    earthquake_data[['latitude', 'longitude']],
    name='Heatmap',
    min_opacity=0.2,
    radius=10,  # Add a comma after this line
    blur=15,
    max_zoom=1,
).add_to(m)

# Display the map
m
```

```python
import folium
import pandas as pd

# Create a list of latitude and longitude coordinates
coords = [
    [47.516232, 14.550072],
    [41.871941, 12.567380],
    [45.943161, 24.966761],
    [50.503887, 4.469936],
]

# Create a DataFrame from the coordinates
data = pd.DataFrame(coords, columns=['latitude', 'longitude'])

# Create a map centered on the first data point
m = folium.Map(location=[data.iloc[0]['latitude'], data.iloc[0]['longitude']], zoom_start=10)

# Add markers to the map for each data point
for _, row in data.iterrows():
    folium.Marker(location=[row['latitude'], row['longitude']]).add_to(m)

# Display the map
m
```

```python
import folium
import pandas as pd

# Create a DataFrame from the coordinates
data = pd.DataFrame(coords, columns=['latitude', 'longitude'])

# Create a map cenetered on the first data point
m = folium.Map(location=[data.iloc[0]['latitude'], data.iloc[0]['longitude']], zoom_start=10)

# Add markers to the map for each data point
for i, row in data.iterrows():
    color = 'red' if i== 0 else 'blue' if i == 1 else 'green' if i == 2 else 'orange' if i == 3 else 'purple'
    folium.Marker(location=[row['latitude'], row['longitude']], icon=folium.Icon(color=color)).add_to(m)

# Display the map
m
```

```python
import folium
import pandas as pd

# Create a DataFrame from the coordinates
data = pd.DataFrame(coords, columns=['latitude', 'longitude'])

# Create a map cenetered on the first data point
m = folium.Map(location=[data.iloc[0]['latitude'], data.iloc[0]['longitude']], zoom_start=10)

# Create a heatmap layer from the data
heatmap_data = data[['latitude', 'longitude']].values.tolist()
folium.plugins.HeatMap(heatmap_data).add_to(m)

# Display the map
m
```
